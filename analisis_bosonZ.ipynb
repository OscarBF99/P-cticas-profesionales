```python
df = pd.read_csv(r"/content/Zmumu.csv", delimiter=',', header=0)
```

```python
# Medidas de tendencia central (pt1,pt2,phi1,phi2)

media_pt1 = df['pt1'].mean()
print("Media de pt1:", media_pt1)
media_pt2 = df['pt2'].mean()
print("Media de pt2:", media_pt2)
media_phi1 = df['phi1'].mean()
print("Media de phi1:", media_phi1)
media_phi2 = df['phi2'].mean()
print("Media de phi2:", media_phi2)

mediana_pt1 = df['pt1'].median()
print("Mediana de pt1:", mediana_pt1)
mediana_pt2 = df['pt2'].median()
print("Mediana de pt2:", mediana_pt2)
mediana_phi1 = df['phi1'].median()
print("Mediana de phi1:", mediana_phi1)
mediana_phi2 = df['phi2'].median()
print("Mediana de phi2:", mediana_phi2)

moda_pt1 = df['pt1'].mode()[0]
print("Moda de pt1:", moda_pt1)
moda_pt2 = df['pt2'].mode()[0]
print("Moda de pt2:", moda_pt2)
moda_phi1 = df['phi1'].mode()[0]
print("Moda de phi1:", moda_phi1)
moda_phi2 = df['phi2'].mode()[0]
print("Moda de phi2:", moda_phi2)
```

```python
# Medidas de dispersion (pt1,pt2,phi1,phi2)

varianza_pt1 = df['pt1'].var()
print("Varianza de pt1:", varianza_pt1)
varianza_pt2 = df['pt2'].var()
print("Varianza de pt2:", varianza_pt2)
varianza_phi1 = df['phi1'].var()
print("Varianza de phi1:", varianza_phi1)
varianza_phi2 = df['phi2'].var()
print("Varianza de phi2:", varianza_phi2)

desviacion_pt1 = df['pt1'].std()
print("Desviacion estandar pt1:", desviacion_pt1)
desviacion_pt2 = df['pt2'].std()
print("Desviacion estandar pt2:", desviacion_pt2)
desviacion_phi1 = df['phi1'].std()
print("Desviacion estandar phi1:", desviacion_phi1)
desviacion_phi2 = df['phi2'].std()
print("Desviacion estandar phi2", desviacion_phi2)
```

```python
rango_pt1 = df['pt1'].max() - df['pt1'].min()
print("Rango de pt1:", rango_pt1)
rango_pt2 = df['pt2'].max() - df['pt2'].min()
print("Rango de pt2:", rango_pt2)
rango_phi1 = df['phi1'].max() - df['phi1'].min()
print("Rango de phi1:", rango_phi1)
rango_phi2 = df['phi2'].max() - df['phi2'].min()
print("Rango de phi2:", rango_phi2)

cuartiles_pt1 = df['pt1'].quantile([0.25,0.50,0.75])
print("cuartiles de pt1:", cuartiles_pt1)
cuartiles_pt2 = df['pt2'].quantile([0.25,0.50,0.75])
print("cuartiles de pt2:", cuartiles_pt2)
cuartiles_phi1 = df['phi1'].quantile([0.25,0.50,0.75])
print("cuartiles de phi1:", cuartiles_phi1)
cuartiles_phi2 = df['phi2'].quantile([0.25,0.50,0.75])
print("cuartiles de phi2:", cuartiles_phi2)
```

```python
# Filtrar de forma que solo se tengan eventos con pT > 20 (para ambas partículas, muón 1 y muón 2)

mask_pt = (df["pt1"] > 20) & (df["pt2"] > 20)
print("Eventos totales:", len(df))
print("Eventos filtrados > 20 :", mask_pt.sum())

print(df[["pt1","pt2"]].dtypes)
print(df[["pt1","pt2"]].head())
```

```python
# Filtrar para obtener eventos con |eta|<2.1 (para ambos, eta1 y eta2)

mask_eta = (df["eta1"].abs() < 2.1) & (df["eta2"].abs() < 2.1)
print("Eventos totales:", len(df))
print("Eventos filtrados < 2.1 :", mask_eta.sum())

print(df[["eta1","eta2"]].dtypes)
print(df[["eta1","eta2"]].head())
```

```python
# Filtrar para obtener eventos con iso<2.5 (para ambos, iso1 e iso2).

mask_pt = (df["iso1"] < 2.5) & (df["iso2"] < 2.5)
print("Eventos totales:", len(df))
print("Eventos filtrados < 2.5 :", mask_pt.sum())

print(df[["iso1","iso2"]].dtypes)
print(df[["iso1","iso2"]].head())
```

```python
# Usando las variables de la muestra, calcular y agregar una nueva columna llamada "M"


df["M"] = M = np.sqrt(2 * df["pt1"] * df["pt2"] * (np.cosh(df["eta1"] - df["eta2"]) - np.cos(df["phi1"] - df["phi2"])))

# Verificamos la nueva columna

print("Masa invariante :")
print(df["M"].dtypes)
print(df["M"].head(10))
```

```python
# Aplicar los filtros para crear un DataFrame filtrado

mask_pt = (df["pt1"] > 20) & (df["pt2"] > 20)
mask_eta = (df["eta1"].abs() < 2.1) & (df["eta2"].abs() < 2.1)
mask_iso = (df["iso1"] < 2.5) & (df["iso2"] < 2.5)

df_filtrado = df[mask_pt & mask_eta & mask_iso].copy()
print(mask_pt.head())

# Lista de variables que queremos comparar
variables = ['pt1', 'eta1', 'iso1']
colores = ['red', 'blue', 'green']

# Crear una figura con 3 sub-gráficas (una al lado de la otra)
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

for i, var in enumerate(variables):
    # 1. Graficar datos originales (Antes)
    axs[i].hist(df[var], bins=50, alpha=0.4, label='Antes (Original)', color='gray')

    # 2. Graficar datos filtrados (Después)
    axs[i].hist(df_filtrado[var], bins=50, alpha=0.8, label='Después (Filtrado)', color=colores[i])

    if var == 'iso1':
        axs[i].set_yscale('log')

    # Personalización de cada sub-gráfica
    axs[i].set_title(f'Comparación de {var}')
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Eventos')
    axs[i].legend()
    axs[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Gráfica : MOMENTO TRANSVERSO (pt2)
plt.figure(figsize=(8, 5))
plt.hist(df['pt2'], bins=100, range=(0, 150), alpha=0.4, label='Antes (Original)', color='gray')
plt.hist(df_filtrado['pt2'], bins=100, range=(0, 150), alpha=0.8, label='Después (Filtrado)', color='orange')
plt.title('Distribución de pt2: Efecto del Filtro > 20 GeV')
plt.xlabel('pt2 [GeV]')
plt.ylabel('Número de Eventos')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Gráfica : PSEUDORRAPIDEZ (eta2)
plt.figure(figsize=(8, 5))
plt.hist(df['eta2'], bins=100, range=(-3, 3), alpha=0.4, label='Antes (Original)', color='gray')
plt.hist(df_filtrado['eta2'], bins=100, range=(-3, 3), alpha=0.8, label='Después (Filtrado)', color='purple')
plt.title('Distribución de eta2: Efecto del Filtro |eta| < 2.1')
plt.xlabel('eta2')
plt.ylabel('Número de Eventos')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Gráfica : AISLAMIENTO (iso2)
plt.figure(figsize=(8, 5))
plt.hist(df['iso2'], bins=100, range=(0, 20), alpha=0.4, label='Antes (Original)', color='gray')
plt.hist(df_filtrado['iso2'], bins=100, range=(0, 20), alpha=0.8, label='Después (Filtrado)', color='cyan')
plt.title('Distribución de iso2: Filtro < 2.5')
plt.xlabel('iso2')
plt.ylabel('Número de Eventos')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# 1. Aplicamos los filtros en orden secuencial
n_total = len(df)

# Corte 1: pT > 20
mask_pt = (df['pt1'] > 20) & (df['pt2'] > 20)
df_paso1 = df[mask_pt]
n_pt = len(df_paso1)

# Corte 2: |eta| < 2.1
mask_eta = (df_paso1['eta1'].abs() < 2.1) & (df_paso1['eta2'].abs() < 2.1)
df_paso2 = df_paso1[mask_eta]
n_eta = len(df_paso2)

# Corte 3: iso < 2.5
mask_iso = (df_paso2['iso1'] < 2.5) & (df_paso2['iso2'] < 2.5)
df_paso3 = df_paso2[mask_iso]
n_iso = len(df_paso3)

# Tabla Cutflow
data_flujo = {
    'Corte aplicado': ['Sin filtros', 'pT > 20 GeV', '|eta| < 2.1', 'iso < 2.5'],
    'Eventos Restantes': [n_total, n_pt, n_eta, n_iso],
    '% Eficiencia (acum.)': [100, (n_pt/n_total)*100, (n_eta/n_total)*100, (n_iso/n_total)*100]
}

tabla_flujo = pd.DataFrame(data_flujo)
print("TABLA DE FLUJO DE CORTES")
print(tabla_flujo)
# Corte más y menos significativo (por pérdida absoluta)
perdida_pt = n_total - n_pt
perdida_eta = n_pt - n_eta
perdida_iso = n_eta - n_iso

perdidas = {'pt': perdida_pt, 'eta': perdida_eta, 'iso': perdida_iso}
mas_sig = max(perdidas, key=perdidas.get)
menos_sig = min(perdidas, key=perdidas.get)

print(f"\nEl corte MÁS significativo es: {mas_sig} (eliminó {perdidas[mas_sig]} eventos)")
print(f"El corte MENOS significativo es: {menos_sig} (eliminó {perdidas[menos_sig]} eventos)")
```

```python
# Gráfica de la masa invariante

plt.figure(figsize=(9, 6))
plt.hist(df["M"], bins=80, range= (0, 200 ), alpha=0.3,
         label="Antes de cortes (Fondo)", color="gray")
plt.hist(df_filtrado["M"], bins=80, range=(0, 200), alpha=0.3,
         label="Después de cortes (Señal Z)", color="green", histtype='stepfilled')
plt.title("Compararción de las Masas invariantes despueés de los cortes")
plt.xlabel(" Masa invariante M [GeV]")
plt.ylabel('Eventos por bin')
plt.axvline(91.18, color='red', linestyle='--', linewidth=1.5, label='Masa Teórica Z (91.18 GeV)')

plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Valores de los picos

promedio_antes = df["M"].mean()
pico_frecuente = df["M"].round(0).mode()[0]

print("Valores del Pico (Bosón Z)")
print(f"Valor del pico (Moda):     {pico_frecuente:.2f} GeV")
print(f"Promedio de la campana:    {promedio_antes:.2f} GeV")
print(f"Desviación estándar:       {df['M'].std():.2f} GeV")


promedio_después = df_filtrado["M"].mean()
pico_frecuente = df_filtrado["M"].round(0).mode()[0]

print("Valores del Pico (Bosón Z)")
print(f"Valor del pico (Moda):     {pico_frecuente:.2f} GeV")
print(f"Promedio de la campana:    {promedio_después:.2f} GeV")
print(f"Desviación estándar:       {df_filtrado['M'].std():.2f} GeV")
```
